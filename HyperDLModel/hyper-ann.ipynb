# Load Data 




```python
data_train = pd.read_excel('/kaggle/input/spam-review-detection/data_train.xlsx')
data_dev = pd.read_excel('/kaggle/input/spam-review-detection/data_dev.xlsx')
data_test = pd.read_excel('/kaggle/input/spam-review-detection/data_test.xlsx')
```

```python
y_train=data_train['Label']

y_dev=data_dev['Label']

y_test=data_test['Label']
```

## CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=100)
```

```python
x_train_count=cv.fit_transform(data_train['transformed_text']).toarray()
x_dev_count=cv.transform(data_dev['transformed_text']).toarray()
x_test_count=cv.transform(data_test['transformed_text']).toarray()
x_train_count.shape
```

# Build Hyper Model

## ANN CountVectorizer

```python
from tensorflow.keras.layers import Dense, Input, Flatten, SimpleRNN, LSTM
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
# import tensorflow_addons as tfa
```

```python
def model_builder (hp):
  model_ANN = tf.keras.Sequential() 
  model_ANN.add(tf.keras.layers.Dense(
                  units = hp.Int('units0', min_value = 32, max_value = 512, step = 32), 
                  activation=hp.Choice('activation0', ['relu', 'tanh', 'sigmoid', 'softmax']),
                  input_dim = x_train_count.shape[1]))
  # Dropout
  model_ANN.add (tf.keras.layers.Dropout(hp.Float(f'dropout1', min_value = 0.1, max_value = 0.3, step = 0.05)))
  # loop
  for i in range (hp.Int('num_layers', 1, 8)):
    model_ANN.add(tf.keras.layers.Dense(units=hp.Int(f'units_{i}', min_value=32, max_value=1024, step=32),
                                        activation=hp.Choice(f'activation_{i}', ['relu', 'tanh', 'sigmoid', 'softmax'])
                                       ))

    model_ANN.add(tf.keras.layers.Dropout(hp.Float(f'dropout_{i}', min_value=0.1, max_value=0.3, step=0.05)))

# Batch size 
  batch_size = hp.Int ('batch_size', 16, 128, step=16)

  model_ANN.add(tf.keras.layers.Dense(units = 1))

# Learning rate
  lr = hp.Float ('lr', 1e-4, 1e-2)
# Epochs
  num_epochs = hp.Int('num_epochs', min_value=5, max_value=20, step=5)
# Compile
  model_ANN.compile(
      loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
      optimizer=Adam(learning_rate=lr),
      metrics=['accuracy']
  )
  return model_ANN
```

```python
# Khởi tạo tuner
# tuner = kt.Hyperband (model_builder, 
#                       objective = 'val_accuracy', 
#                       max_epochs=10,
#                       factor=3, 
#                       hyperband_iterations=0,
#                       directory='my_dir', 
#                       project_name='my_project')

tuner = kt.Hyperband(
    model_builder,
    objective='val_accuracy',
    max_epochs = 10,
    factor = 3,
    directory='my_dir',
    project_name='my_project'
)
!rm -rf /content/tuner_dir_1
# max_epochs : Integer, the maximum number of epochs to train one model. It is recommended to set this to a value slightly higher than the expected epochs to convergence for your largest Model
# factor: Integer, the reduction factor for the number of epochs and number of models for each bracket. Defaults to 3.
```

```python
#from kerastuner import Hyperband as hp
hp = kt.HyperParameters()
# hp.Int('batch_size', 16, 128, step=16)

# The best HyperParameter 
tuner.search (x_train_count, 
              y_train, 
              epochs = 10,               
              validation_data = (x_dev_count, y_dev), 
              batch_size=hp.Int('batch_size', 16, 128, step=16), 
#               activation= activation,
              callbacks = [early_stop])
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(best_hps.values)
```

```python
print('Number of hidden layers:', best_hps.get('num_layers'))

for i in range(best_hps.get('num_layers')):
    print(f'Units for layer {i+1}:', best_hps.get(f'units_{i}'))
    print(f'Dropout for layer {i+1}:', best_hps.get(f'dropout_{i}'))
    if 'activation0' in best_hps:
        activation_fn = best_hps.get(f'activation_{i}')
        print(f'Activation Function for layer {i+1}:', activation_fn)
    else:
        print('Activation parameter does not exist.')
        
# print(best_hps.get("activation0"))
print('Learning rate:', best_hps.get('lr'))
print('Batch size:', best_hps.get('batch_size'))
print('Number of epochs:', best_hps.get('num_epochs'))
# print('Best Activavtion Function:', best_hps.get('activation{i}'))
```

## Build Model with the best HP

```python
best_model = tuner.get_best_models ()[0]
best_model.build (x_train_count.shape)
best_model.summary()
```

```python
best_Model = tuner.hypermodel.build (best_hps)
```

```python
best_Model.fit(x_train_count, 
                y_train, 
                epochs = 10, 
                validation_data = (x_dev_count, y_dev), 
                batch_size=32,
                )
```

```python
loss, accuracy = best_Model.evaluate(x_test_count, y_test)
```
